```python
#Curve Fitting
np.random.seed(101)
x = np.linspace(0.1,3,20).round(2)
x
y=np.round(np.sin(x)+0.1*np.random.randn(20),2)#Adding random noise#If no noise data is clean
y
# p=np.poly1d(np.polyfit(x,y,15))#Overfitting
p=np.poly1d(np.polyfit(x,y,3))#Model
print(p)
# p is polynomial/Poly1d extracts the variables from polyfit.polyfit makes it an equation
print(p),p(0.5)
p([0.5,0.6,0.7])
#We are fitting this p curve fitting.We are capturing the pattern.
```

```python
fig=plt.figure(figsize=(10,5))#10/5 inches #This is to create the plot axis and size is provided as tuple

plt.plot(x,y,'bo',label='Original Data')#color of the line & markers,linestyle,markerstyle#This part of label comes in legend
t=np.linspace(0.1,3,100)
plt.plot(x,p(x),'r-',label='Poly Fit')
plt.xlabel('X values $3\pi$',fontsize=18)#These are the labels.Show the x axis
#$3\pi$  3pi will be shown  ,\This gives pi symbol
plt.ylabel('Y values $\Delta^{2\pi}$',fontsize=18)#^ This makes it power
plt.title('Curve Fitting $\mu_H$',fontsize=21)#_ is subscript
plt.legend(fontsize=15,loc='lower center')
# plt.xticks(xticks)
plt.grid(True)

plt.show()#Will remove the detail
#Always save the plot after it has been rendered or shown

fig.savefig('name of the figure.png',dpi=200)#dip is not mandatory it is for resolution
```

## Evaluation matrices for the polynomial.

## The best-fitting polynomial will have lowest MSE (lower limit of 0),max R2(upper limit of 1)

To check fitness of a polynomial

```python
#You can use Mean Square Error(MSE) or R2

y_pred=p(x)
error=y_pred-y
MSE=(error**2).mean()
MSE
```

```python
# (Coefficient of determination)R2 or (Goodness of Fit)R2
# SSR(Sum of Square of residuals)
# SST(Sum of squares of the total)
SSR=(error**2).sum()#Residuals is same as error
SST=((y-y.mean())**2).sum()#sum of squares of deviations of y from its mean
R2=1-(SSR/SST)
R2
```

```python
from sklearn.metrics import r2_score,mean_squared_error
r2_score(y,y_pred)
mean_squared_error(y,y_pred)
```

```python
x=np.linspace(0,10,100)
y=2*x - 5 + 0.1*np.random.rand(100)
plt.plot(x,y,'r-')
```

```python
#Elbow Method degree selection.Maximum perpendicular distance.
MSE_val=[]
R2_val=[]
for deg in range(1,15):
    p=np.poly1d(np.polyfit(x,y,deg))
    MSE_val.append(mean_squared_error(y,p(x)))
    R2_val.append(100*r2_score(y,p(x)))
```

```python
degrees=list(range(1,15))
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(degrees,MSE_val,'b-o')
plt.ylabel('MSE Values')
plt.grid(True)
plt.subplot(1,2,2)
plt.plot(degrees,R2_val,'r-o')
plt.ylabel('R2 Values')
```
